```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(0)  # For reproducibility
work_experience = np.random.randint(1, 21, 100)  # Work experience (1 to 20 years)
salaries = 2000 + 1500 * work_experience + np.random.normal(0, 10000, 100)  # Salary with some noise

# Reshape data for sklearn
X = work_experience.reshape(-1, 1)  # Independent variable
y = salaries  # Dependent variable

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)  # Fit the model

# Predictions
predictions = model.predict(X)

# Plotting
plt.scatter(X, y, color='blue', label='Actual Data')  # Scatter plot of actual data
plt.plot(X, predictions, color='red', linewidth=2, label='Regression Line')  # Regression line
plt.title('Work Experience vs Salary')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Salary (USD)')
plt.legend()
plt.grid()
plt.show()

# Coefficients
B0 = model.intercept_  # Intercept
B1 = model.coef_[0]  # Slope

# Output coefficients
print(f'Intercept (B0): {B0}')
print(f'Slope (B1): {B1}')
```
